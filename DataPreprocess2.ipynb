```python
import cv2
import numpy as np
from tensorflow.keras.preprocessing import image
import glob

# Set the path to the directory containing your brain MRI images
image_directory = '/CSE499/Dataset/Non_Demented/'

# Get a list of all image file paths in the directory
image_paths = glob.glob(image_directory + '*.jpg')

# Initialize an empty array to store the processed image data
image_data = []

# Loop through the image paths and process each image
for image_path in image_paths:
    # Load the image using keras.preprocessing
    img = image.load_img(image_path)
    
    # Convert the image to a NumPy array
    img_array = image.img_to_array(img)
    
    # Append the processed image array to the dataset
    image_data.append(img_array)

# Convert the list of image arrays to a NumPy array
image_data = np.array(image_data)

# Save the dataset as a .npy file
np.save('/CSE499/Dataset/ConvertedDataset/NonDemented.npy', image_data)
```
